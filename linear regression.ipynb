```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([300,500,700,900,1100])
```

```python
def cost_fn(x=x,y=y):
    w_arr=np.linspace(-100,300,100) 
    b_arr=np.linspace(-300,300,100)
    m=x.size
    j_arr=np.zeros((len(w_arr),len(b_arr)))

    for i,w in enumerate(w_arr):
        for j,b in enumerate(b_arr):
            f_x=w*x+b
            cost=np.sum((f_x-y)**2)/(2*m)
            j_arr[i,j]=cost
            
    b,w=np.meshgrid(b_arr,w_arr)
    cp=plt.contourf(b,w,j_arr,levels=100)
    plt.xlabel("b values")
    plt.ylabel("w values")
    plt.title("cost fn")
    plt.colorbar(cp)
    plt.show()
    min_cost = np.min(j_arr)
    print("Minimum cost:", min_cost)
    
    
```

```python
cost_fn(x,y)
```

```python
def gradient_des(x,y):
    alpha=1e-2
    m=x.size
    n=70
    w=0
    b=0
    j=[]
    for i in range(n):
        f_x=w*x+b
        dj_dw=np.sum((f_x-y)*x)/m
        dj_db=np.sum(f_x-y)/m
        w=w-alpha*dj_dw
        b=b-alpha*dj_db
        cost=np.sum((f_x-y)**2)/(2*m)
        j.append(cost)
    plt.figure(figsize=(5,5))
    plt.plot(range(n),j)
    plt.xlabel("Iterations")
    plt.ylabel("Cost")
    plt.title("Cost Function vs Iterations")
    plt.show()
    return w,b,j
```

```python
w,b,j=gradient_des(x,y)
plt.figure(figsize=(5.23,5))
plt.scatter(x,y,label='data')
plt.plot(x,w*x+b,label='fitted line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear regression')
plt.legend()
plt.show()
```
